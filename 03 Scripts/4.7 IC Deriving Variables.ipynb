#### 3. Define price_label function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

#### 4. Apply price_label function to dataframe and check results

```python
ords_prods_merge['price_range'] = ords_prods_merge.apply(price_label, axis=1)
```

```python
ords_prods_merge['price_range'].value_counts(dropna = False)
```

#### 5. Create same function using loc()

```python
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['price_range'].value_counts(dropna=False)
```

#### 6. Create new column using For loop

```python
result = []

for value in ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
###### 7.2 Create a new column for labeling the 2 busiest days of the week as busiest_days using a suitable method
```

```python
# check what days are the busiest
ords_prods_merge['order_day_of_week'].value_counts(dropna = False) 
```

```python
# create new column
ords_prods_merge.loc[ords_prods_merge['order_day_of_week'].isin([0, 1]), 'busiest_days'] = 'Busiest Days'
ords_prods_merge.loc[ords_prods_merge['order_day_of_week'].isin([3, 4]), 'busiest_days'] = 'Slowest Days'
ords_prods_merge.loc[ords_prods_merge['order_day_of_week'].isin([6, 2, 5]), 'busiest_days'] = 'Regular Days'
```

```python
###### 7.4 Create a new column busiest_period_of_day with labels Most orders, Average orders and Fewest orders
```

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create a list with new labels
```

```python
result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12:
    result.append("Most orders")
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
result
```

```python
# Assign new list to new column
```

```python
ords_prods_merge['busiest_period_of_day'] = result
```
